```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from preprocess import preprocess
```

```python
df_init = pd.read_csv("gmpe.csv")
df = preprocess(df_init)
df.head()
```

```python
num_cols = df.drop(["FaultType_Ters","FaultType_Yanal"],axis=1).columns
#ss = StandardScaler()
#X_scaled = StandardScaler().fit_transform(df[num_cols])
#df[num_cols] = pd.DataFrame(X_scaled, columns=df[num_cols].columns)
x_train, x_test, y_train, y_test = train_test_split(df.drop("PGA", axis=1), df["PGA"], test_size=0.2, random_state=42)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units = 5, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 10, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 5, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 1))
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(x_train.values, y_train.values, batch_size = 32, epochs = 100)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test.values.reshape(len(y_test),1), y_pred)
```
